```python
#Task 1
import numpy as np
```

```python
#Task 2
temps = np.array([57, 55, 54, 53, 53, 54, 56, 60, 65, 70, 74, 78, 81, 83, 84, 84, 82, 79, 74, 70, 66, 63, 61, 59])
print(temps)
print(temps.shape)
print(temps.dtype)
```

```python
#Task 3
temps_c = (temps - 32) * 5 / 9
print(temps_c)
```

```python
#Task 4
print("Sum: ", np.sum(temps_c))
print("Mean: ", np.mean(temps_c))
print("Min: ", np.min(temps_c))
print("Max: ", np.max(temps_c))
print("Median: ", np.median(temps_c))
```

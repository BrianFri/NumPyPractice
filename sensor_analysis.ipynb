```python
#Task 1
import numpy as np
```

```python
#Task 2
temps = np.array([57, 55, 54, 53, 53, 54, 56, 60, 65, 70, 74, 78, 81, 83, 84, 84, 82, 79, 74, 70, 66, 63, 61, 59])
print(temps)
print(temps.shape)
print(temps.dtype)
```

```python
#Task 3
temps_c = (temps - 32) * 5 / 9
print(temps_c)
```

```python
#Task 4
print("Sum: ", np.sum(temps_c))
print("Mean: ", np.mean(temps_c))
print("Min: ", np.min(temps_c))
print("Max: ", np.max(temps_c))
print("Median: ", np.median(temps_c))
```

```python
#Task 5
print(f"2:00 PM temperature: {temps[14]}°F")
```

```python
#Task 6
afternoon = temps[12: 18]
print(afternoon)
```

```python
#Task 7
# Hot Hours
hot_hours = temps[temps > 80]
print(hot_hours)
print("Number of hot hours: ", hot_hours.size)
```

```python
#Task 8
sensor_grid = np.array([[81, 83, 84, 84, 82, 79],
                       [78, 80, 82, 83, 81, 77],
                       [83, 85, 87, 88, 86, 82]])
print(sensor_grid)
print(sensor_grid.shape)
```

```python
#Task 9
print("Sensor 3 hour 3 reading: ", sensor_grid[2, 3])
print("Sensor 2 readings", sensor_grid[1, :])
print("All senor readings for hour 1", sensor_grid[:, 0])
```

```python
#Task 10
sensor_1 = np.mean(sensor_grid[0,:])
print(sensor_1)
sensor_2 = np.mean(sensor_grid[1,:])
print(sensor_2)
sensor_3 = np.mean(sensor_grid[2,:])
print(sensor_3)
```

```python
#Task 11
range_first_day = np.max(sensor_grid[0,:]) - np.min(sensor_grid[0,:])
range_second_day = np.max(sensor_grid[1,:]) - np.min(sensor_grid[1,:])
range_third_day =  np.max(sensor_grid[2,:]) - np.min(sensor_grid[2,:])

hourly_range = np.array([range_first_day, range_second_day, range_third_day])
```

```python
print(hourly_range)
```
